#### 1. Explain the theoretical **Simple Linear Regression** model in your own words by describing its components (of predictor and outcome variables, slope and intercept coefficients, and an error term) and how they combine to form a sample from **normal distribution**; then, (working with a ChatBot if needed) create `python` code explicitly demonstrating your explanation using `numpy` and `scipy.stats`<br><br>

simple linear regression is a statistic method used to decribe the linear relationship between the independent and dependent variables

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Define parameters
n = 50            # Number of data points
beta0 = 1.0       # True intercept
beta1 = 2.5       # True slope
sigma = 2.0       # Standard deviation of the noise

# Generate the independent variable x
np.random.seed(42)  # For reproducibility
x = np.linspace(0, 10, n)

# Generate errors with standard deviation sigma
errors = np.random.normal(0, sigma, n)

# Calculate the dependent variable y with a linear relationship and added noise
y = beta0 + beta1 * x + errors

# Perform simple linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Print the results
print("Slope:", slope)
print("Intercept:", intercept)
print("R-squared:", r_value**2)
print("P-value:", p_value)
print("Standard error:", std_err)

# Plot the data points and the regression line
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, slope * x + intercept, color='red', label='Regression line')
plt.xlabel("Independent Variable (x)")
plt.ylabel("Dependent Variable (y)")
plt.legend()
plt.title("Simple Linear Regression Example with Specified Parameters")
plt.show()
````

#### 2. Continuing the previous question... (working with a ChatBot if needed) use a dataset **simulated** from your theoretical **Simple Linear Regression** model to demonstrate how to create and visualize a fitted **Simple Linear Regression** model using `pandas` and `import statsmodels.formula.api as smf`<br><br>

first is to create x and y with normally distributed noise, which are the random deviations from the normal distributions. then we store the x and y to DataFrame which allows us to leverage formula-based modeling(bystoring the data in DataFrame, statsmodels allows us to use formulas to specify relationships with variables). after that, we use smf.ols() to specify the formula Y~X, where X is independent variable and Y is dependent variable. based on this model, using .fit() to perform the regression analysis. use model.summary() to provide detailed output included key elements like coefficients, p-values and standard errors. and last is to use model.predict(data) to do plotting for well visualization.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Set random seed for reproducibility
np.random.seed(0)

# Define parameters for the linear model
beta_0 = 2  # Intercept
beta_1 = 0.5  # Slope
n_samples = 100  # Number of data points
sigma = 1  # Standard deviation of error term

# Generate random predictor variable (independent variable X)
X = np.linspace(0, 10, n_samples)

# Generate error term (normally distributed noise)
error = np.random.normal(0, sigma, n_samples)

# Generate outcome variable (dependent variable Y) using the regression equation
Y = beta_0 + beta_1 * X + error

# Create a pandas DataFrame from the data
data = pd.DataFrame({'X': X, 'Y': Y})

# Fit a Simple Linear Regression model using statsmodels
model = smf.ols(formula='Y ~ X', data=data).fit()

# Print the summary of the regression model
print(model.summary())

# Visualization
plt.figure(figsize=(10, 6))

# Scatter plot of the original data points
plt.scatter(data['X'], data['Y'], color='blue', label='Data Points')

# Plot the fitted regression line
plt.plot(data['X'], model.predict(data), color='red', label='Fitted Regression Line')

# Add labels and title
plt.xlabel("Predictor Variable (X)")
plt.ylabel("Outcome Variable (Y)")
plt.title("Fitted Simple Linear Regression Model")
plt.legend()
plt.show()
````

As seen in the introductory figure above, if the delay of the geyser eruption since the previous geyser eruption exceeds approximately 63 minutes, there is a notable increase in the duration of the geyser eruption itself. In the figure below we therefore restrict the dataset to only short wait times. Within the context of only short wait times, is there evidence in the data for a relationship between duration and wait time in the same manner as in the full data set? Using the following code, characterize the evidence against the **null hypothesis** in the context of short wait times which are less than  `short_wait_limit` values of `62`, `64`, `66`.<br><br>

````python
import pandas as pd
import seaborn as sns
import statsmodels.formula.api as smf

# Load the Old Faithful Geyser dataset
old_faithful = sns.load_dataset('geyser')

# Define the different limits for short wait times
short_wait_limits = [62, 64, 66]

# Function to fit and summarize a model for each short wait limit
def analyze_short_wait_data(limit):
    # Subset data for waiting times less than the given limit
    short_wait_data = old_faithful[old_faithful['waiting'] < limit]
    # Define and fit the linear model
    model = smf.ols('duration ~ waiting', data=short_wait_data).fit()
    # Print the summary of the model for interpretation
    print(f"Results for waiting times less than {limit} minutes:")
    print(model.summary())
    print("\n")

# Run analysis for each short wait limit
for limit in short_wait_limits:
    analyze_short_wait_data(limit)
````

-if the p-value of the waiting time stays below 0.05, the waiting and duration has shorter wait times. and if having a high p-value, the waiting time and eruption duration may be weaken in the restricted ranges.

-if the coefficient value is large and positive, the waiting time will become short, and if the value is small, the relationship will weaken as the waiting time getting short with the observed increase in eruption duration for longer waiting times.

-a lower R^2 suggest that a shorter waiting time itself cannot make interpretations that the variation in eruption duration. therefore it could support the idea that waiting time has a stronger impact when it exceeds a certain threshold

to conclude, this analysis helps to assess whether shorter wait times affect eruption duration in a similar way that is observed in the full dataset. If significant, which means that short waits predicted longer eruptions; if not, it could indicate that the relationship is primarily driven by longer waits.
