Difference between Attributes and Methods:
Attribute: A property of an object accessed without parentheses. Example: df.shape returns the dimensions of the DataFrame.
Method: A function associated with an object, accessed with parentheses. Example: df.describe() returns statistical summaries of numerical columns.
Statistical Measures from df.describe():
Count: Number of non-null entries.
Mean: Average value of the data.
Std: Standard deviation, indicating the spread of data.
Min: Minimum value in the column.
25%: 25th percentile (first quartile), below which 25% of the data falls.
50%: 50th percentile (median), middle value of the data.
75%: 75th percentile (third quartile), below which 75% of the data falls.
Max: Maximum value in the column.
Use Cases for df.dropna() vs. del df['col']:
Using df.dropna(): Preferred when you want to remove rows with missing values in specific columns but keep the column. Example code:

```python
import pandas as pd

# Sample DataFrame
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Email': ['alice@example.com', None, 'charlie@example.com', None]
})

# Remove rows where 'Email' column is NaN
df_cleaned = df.dropna(subset=['Email'])
print(df_cleaned)
```

Using del df['col']: Preferred when you want to remove an entire column. Example code:

```python
# Remove the 'Email' column entirely
del df['Email']
print(df)
```

```python
import pandas as pd

# Example 1: Checking shape and describe
url = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-05-05/villagers.csv"
df = pd.read_csv(url)

# Shape of DataFrame
rows, columns = df.shape
print(f"Number of rows: {rows}")
print(f"Number of columns: {columns}")

# Describe the DataFrame
summary = df.describe()
print("Summary of numerical data:\n", summary)

# Overview of the DataFrame
overview = df.info()
print("\nOverview of the dataset:\n", overview)

# Example 2: Removing rows with missing values
df_with_na = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Email': ['alice@example.com', None, 'charlie@example.com', None]
})
df_cleaned = df_with_na.dropna(subset=['Email'])
print(df_cleaned)

# Example 3: Removing an entire column
df_with_address = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Email': ['alice@example.com', None, 'charlie@example.com', None],
    'Address': ['123 Lane', '456 Street', '789 Road', '101 Avenue']
})
del df_with_address['Address']
print(df_with_address)
```
